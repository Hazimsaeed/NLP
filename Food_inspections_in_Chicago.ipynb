I Use Python Regular Expressions to identify top-10 most frequent causes of failed food inspections in Chicago.  

```python
import re
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random
import sklearn as sk
import os
import math
os.getcwd()
```

```python
df = pd.read_csv("~/Downloads/NLP/Week1/Assignment1/Food_Inspections.csv", error_bad_lines = False, encoding='latin-1')
df.head(2)

df = df[df.Results =='Fail']
df
```

Now select only the column of interest:

```python
df = df[df.Results =='Fail']
output = ''
for x in df['Violations']:
    output += '|'
    output += str(x).strip()

txt=output.split('|')
```

Clean the text I have and split it in list 

```python
while('' in txt) : 
    txt.remove('')

result=[]    
for x in txt:
    output = str(x).lstrip()
    #output = output.replace('-', '')
    result.append(output) 

txt_clean = [x for x in result if str(x) != 'nan']
   
```

Creat list called cleanedresult has only the text we are interested to count

```python
cleanedresult =[]
find = r"(\d{1,2}\.\s.+?) -"
for x in txt_clean:
    cleanedresult.append(re.findall(find,x))
```

Put the text we are interested in a dataframe so easily we can count the most common and print out the result 

```python
df_count = pd.DataFrame({'text':cleanedresult})
df_count['text'].value_counts().head(10)
```
